### Data Wrangling
#### 1. Change the currency unit "₹"
#### 2. Change the data type ("object") to the type we need

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
df.head()
```

```python
import matplotlib.pyplot as plt

most_main_items = df_op['category_1'].value_counts().head(10).rename_axis('category_1').reset_index(name='counts')
most_sub_items = df_op['category_2'].value_counts().head(10).rename_axis('category_2').reset_index(name='counts')
miscellaneous_items = df_op['category_3'].value_counts().head(10).rename_axis('category_3').reset_index(name='counts')

colors = sns.color_palette('Set1')
fig = plt.figure(figsize=(10, 4))
bars = plt.barh(most_main_items['category_1'], most_main_items['counts'], color=colors)

for bar, count in zip(bars, most_main_items['counts']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, str(count), va='center', ha='left')

plt.xlabel('Counts')
plt.ylabel('Category')
plt.show()
```

```python
# Rating vs Rating Counts
# Do customers tend to leave more reviews for items they like/ don't like?
rating = df_op['rating'].replace('|', np.nan).astype(float)
rating_counts = df_op['rating_count'].str.replace(',','')
rating_counts= rating_counts.fillna(0).astype(int) # Use fillna since there are NaN in the rating_counts
rating_counts
```

```python
# Visualizing Rating vs Rating Counts
fig = plt.figure(figsize=(6, 4))
sns.scatterplot(x = rating, y = rating_counts)
plt.xlabel('Rating')
plt.ylabel('Number of Ratings')
plt.title('Scatter Plot of Rating vs Counts')
```

```python
# Histogram view
fig = plt.figure(figsize=(6, 4))

# Number of bins decided by Sturges' Formula : int(1 + np.log2(len(x)))
sns.histplot(x = rating, bins=11, kde=True, color='orange')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Rating Distribution')
```

##### According to this result. People tend to leave more ratings when they are satisfied with products (~4)

```python
# Calculate rating for each category_3
df_op['rating'] = df_op['rating'].replace('|', np.nan).astype(float)
rating_cat3 = df_op.groupby('category_3')['rating'].agg(['mean', 'min', 'max'])

# Round the mean rating to 2 decimal places
rating_cat3['mean'] = rating_cat3['mean'].round(2)
rating_cat3
```

##### Here, we can observe the same trend, the mean value of most ratings are close to 4

```python
# Generating wordcloud for selected reviews (electronics)
from wordcloud import WordCloud
df_reviews_electronics = df_op[df_op['category_1'] == 'Electronics']
reviews_electronics = ' '.join(df_reviews_electronics['review_content'].dropna().values)
wordcloud_electronics = WordCloud(width=800, height=400, background_color='white', min_font_size=10).generate(reviews_electronics)
fig = plt.figure(figsize=(8, 4))
plt.imshow(wordcloud_electronics)
plt.axis("off")
plt.show()
```

```python
# Generating wordcloud for selected reviews (home&kitchen category with review >= 3.8)
from wordcloud import WordCloud
df_reviews_home_positive = df_op[(df_op['category_1'] == 'Home&Kitchen') & (df_op['rating'] >= 3.8) ]
reviews_home_positive = ' '.join(df_reviews_home_positive['review_content'].dropna().values)
wordcloud_home_postive = WordCloud(width=800, height=400, background_color='white', colormap = 'Set2', min_font_size=10, min_word_length = 2).generate(reviews_home_positive)
plt.figure(figsize=(8, 4))
plt.imshow(wordcloud_home_postive)
plt.axis("off")
plt.show()
```

#### Analyze what customers care most in their review. Things like "good", "nice" are not helpful in this analysis
#### Use the stopword to remove those

```python
# Modifiy selected reviews (home&kitchen category with review >= 3.8) with stopwords
from wordcloud import WordCloud
stopwords = set()
stopwords.update(["good", "nice", "also", "with", "from", "this", "that", "very", "like", "only", "great"])

wordcloud_home_postive = WordCloud(width=800, height=400, background_color='white', colormap = 'Set2', min_font_size=10, min_word_length = 4, stopwords=stopwords).generate(reviews_home_positive)
plt.figure(figsize=(8, 4))
plt.imshow(wordcloud_home_postive)
plt.axis("off")
plt.show()
```

### Data Wrangling
#### 1. Change the currency unit "₹"
#### 2. Change the data type ("object") to the type we need

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
df.head()
```

```python
df.isna().sum() # Always use isna() over isnull()
```

```python
# Some missing values in rating_count.
nan_cols = [i for i in df.columns if df[i].isnull().any()] # This identify the name of the column
rows_with_nan = df[df.isna().any(axis=1)]
rows_with_nan
```

```python
# Duplicate the df for operation. Remove the unwanted info in the operation part
df_op = df.copy().drop(['user_name','review_id','img_link','product_link'], axis=1)
# Replace the NaN with 0
df_op['rating_count'] = df_op['rating_count'].fillna(0)
df_fillna_check = df_op.iloc[282]
df_fillna_check
```

```python
main = df_op[['category_1', 'category_2', 'category_3', 'category_4','product_id']].rename(columns={'category_1' :'Main Category', 'category_2' : 'Sub-Category', 'product_id':'Products'})
product_by_cat = pd.pivot_table(main, index=['Main Category', 'Sub-Category','category_3', 'category_4'], aggfunc='count')
product_by_cat
```

```python
# category_4 is too detail
main = df_op[['category_1', 'category_2', 'category_3','product_id']].rename(columns={'category_1' :'Main Category', 'category_2' : 'Sub-Category', 'category_3' : 'Sub-Sub-Category','product_id':'Products'})
product_by_cat3 = pd.pivot_table(main, index=['Main Category', 'Sub-Category','Sub-Sub-Category'], aggfunc='count')
product_by_cat3
```

```python
import matplotlib.pyplot as plt

most_main_items = df_op['category_1'].value_counts().head(10).rename_axis('category_1').reset_index(name='counts')
most_sub_items = df_op['category_2'].value_counts().head(10).rename_axis('category_2').reset_index(name='counts')
miscellaneous_items = df_op['category_3'].value_counts().head(10).rename_axis('category_3').reset_index(name='counts')

colors = sns.color_palette('Set1')
fig = plt.figure(figsize=(10, 3))
plt.barh(most_main_items['category_1'], most_main_items['counts'], color=colors)
plt.xlabel('Counts')
plt.ylabel('Category')
plt.show()
```
